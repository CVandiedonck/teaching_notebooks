### 2.1. Chargement/Installation d'un paquet classique

L'installation classique d'un paquet repose sur la fonction `install.packages()`. Le paquet `BiocManager()` de Bioconductor est le seul paquet de Bioconductor √† s'installer classiquement.

Avant de tenter de l'installer, nous pouvons v√©rifier que le paquet `BiocManager` est d√©j√† install√© sur adenine dans le r√©pertoire associ√© √† l'environnement.

```r
#cell 8
installed.packages()["BiocManager",c(1,2,3)]
```

Nous pouvons √† pr√©sent charger le paquet avec la fonction `library()`.

```r
#cell 9
library(BiocManager)
```

Si le paquet n'avez pas encore √©t√© install√©, la commande ci-dessus aurait retourn√© une erreur. 
Dans ce cas, vous auriez pu utiliser la commande ci-dessous pour l'installer avec la fonction `install.packages()`.

<div class="alert alert-block alert-warning"><b> Warning</b>.Cependant, il est inutile de r√©installer un paquet d√©j√† install√©.

L'execution conditionnelle ci-dessous solutionne le probl√®me, que le paquet soit d√©j√† install√© ou pas. En effet, cette commande permet de n'installer le paquet que s'il ne l'est pas encore. C'est ainsi que nous vous conseillons d'invoquer toutes les installations de paquets dans vos scripts afin de permettre une portabilit√© entre utilisateurs, en leur √©pargnant la r√©installation de paquets qu'ils ont d√©j√†. *(Nous verrons les √©x√©cutions conditionnelles plus en d√©tail lors de la session 6 de R.)*

```r
#cell 10
if (!requireNamespace("BiocManager"))
  install.packages("BiocManager")
```

#### 3.3.2. Identifier les genes √† +/- 20 kb des sites d'actelylation contenant au moins un SNP avec la fonction `getBM` du paquet `biomaRt`

```r
#cell 32
info <- data.frame(h3k27ac_SNPs_GR@seqnames@values, h3k27ac_SNPs_GR@ranges@start,
                   h3k27ac_SNPs_GR@ranges@start + h3k27ac_SNPs_GR@ranges@width)
```

```r
#cell 33
str(info)
names(info) <- c("chr", "start", "end")
```

```r
#cell 34
mapgenes <- NULL
for (i in 1:nrow(info)) {
  genes <-getBM(attributes = c("hgnc_symbol", "description", "ensembl_gene_id","chromosome_name",
                              "start_position", "end_position", "strand"),
               filters = c("chromosome_name", "start", "end"),
               values = list(gsub("chr", "", info$chr[i]),info$start[i] - 10000, info$end[i] + 10000), mart = ensembl)
  genes$site <- paste0(info$chr[i], ":", info$start[i], "-", info$end[i])
  mapgenes <- rbind(mapgenes, genes)
}
mapgenes[,c(8, 1:7)] # pour redoronner les colonnes et voir les resultats
```
